# **Los viajes de 1 o 2 personas son mas probables de comprar via mobile, que de comprar via internet**

```python
completados = archivo_dos.loc[archivo_dos.booking_complete == 1]
completados
```

```python
completados.purchase_lead.value_counts()
```

```python
internet = completados.loc[completados.sales_channel == 'Internet']
internet
```

```python
internet.purchase_lead.value_counts()
```

```python
mobile = completados.loc[completados.sales_channel == 'Mobile']
mobile
```

```python
mobile.route.unique()
```

```python
mobile.groupby(['booking_origin', 'flight_day'])['num_passengers'].count()
```

```python
mobile.groupby('booking_origin')['route'].count()
```

```python
mobile.groupby('booking_origin')['num_passengers'].sum()
```

```python
mobile.booking_origin.value_counts()
```

```python
mobile.purchase_lead.value_counts()
```
