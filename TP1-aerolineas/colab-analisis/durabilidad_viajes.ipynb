```python
# La cantidad de viajes con length_of_stay = 0 son pocos a la par de todo el dataframe
# Y como el booking_complete no esta completo, se decide descartar para el analisis
archivo_actualizado.loc[archivo_actualizado.length_of_stay == 0]
```

```python
pasajeros_estadias = pasajeros_estadias.loc[pasajeros_estadias.length_of_stay != 0]
pasajeros_estadias
```

```python
cantidad = pasajeros_estadias.groupby(['num_passengers'])['length_of_stay'].count().to_frame().reset_index()
cantidad
```

```python
pasajeros_estadias['num_passengers'].value_counts()
```

```python
cantidad.pivot(values='cant', columns='length_of_stay', index='num_passengers')
```

```python
pasajeros_estadias.groupby(['num_passengers']).agg({'length_of_stay':['max', 'min', 'count']}).reset_index()
```
